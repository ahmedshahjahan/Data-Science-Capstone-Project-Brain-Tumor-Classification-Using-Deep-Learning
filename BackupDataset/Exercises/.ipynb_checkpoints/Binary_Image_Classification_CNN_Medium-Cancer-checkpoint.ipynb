```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
import os
import cv2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
DATADIR_TRAIN = "../train_bird"
DATADIR_VALID = "../validation_bird"
DATADIR_TEST = "../test_bird"
CATEGORIES = ["bird", "duck"]
```

```python
train = ImageDataGenerator(rescale=1/255)
test = ImageDataGenerator(rescale=1/255)

train_dataset = train.flow_from_directory("../train_cancer",
                                          target_size=(50,50),
                                          batch_size = 32,
                                          class_mode = 'binary')
                                         
validation_dataset = test.flow_from_directory("../validation_cancer",
                                          target_size=(50,50),
                                          batch_size =32,
                                          class_mode = 'binary')
```

```python
validation_dataset.class_indices
```

```python
train_dataset.class_indices
```

```python
model = keras.Sequential()

# Convolutional layer and maxpool layer 1
model.add(keras.layers.Conv2D(32,(3,3),activation='relu',input_shape=(50,50,3)))
model.add(keras.layers.MaxPool2D(2,2))

# Convolutional layer and maxpool layer 2
model.add(keras.layers.Conv2D(64,(3,3),activation='relu'))
model.add(keras.layers.MaxPool2D(2,2))

# Convolutional layer and maxpool layer 3
model.add(keras.layers.Conv2D(128,(3,3),activation='relu'))
model.add(keras.layers.MaxPool2D(2,2))

# Convolutional layer and maxpool layer 4
model.add(keras.layers.Conv2D(128,(3,3),activation='relu'))
model.add(keras.layers.MaxPool2D(2,2))

# This layer flattens the resulting image array to 1D array
model.add(keras.layers.Flatten())

# Hidden layer with 512 neurons and Rectified Linear Unit activation function 
model.add(keras.layers.Dense(512,activation='relu'))

# Output layer with single neuron which gives 0 for Cat or 1 for Dog 
#Here we use sigmoid activation function which makes our model output to lie between 0 and 1
model.add(keras.layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#steps_per_epoch = train_imagesize/batch_size

model.fit(train_dataset,
         epochs = 30, batch_size=10,
         validation_data = validation_dataset  
         )
```

```python
# Manually Predicting 
test_path = '../test_cancer'
Cancer=0
NoCancer=0
for i in os.listdir(test_path):
    img = image.load_img(test_path + "//" +i, target_size=(50,50,3))
    plt.imshow(img)
    plt.show()
    X = image.img_to_array(img)
    X = np.expand_dims(X, axis=0)
    images = np.vstack([X])
    val = model.predict(images)
    if val == 0:
        Cancer +=1
        print('Cancer')
    else:
        NoCancer +=1
        print('No Cancer')
print(f'Cancer: {Cancer}, NoCancer: {NoCancer}')
```
